# 1. On-Policy Reinforcement Learning with Parametric Policy

```python
import lqg1d
import matplotlib.pyplot as plt
import utils
```

```python
class GaussianPolicy:
    
    def __init__(self, theta, w, mean, var, grad_mean, grad_var):
        self.init_theta = theta
        self.init_w = w
        self.mean = mean
        self.var = var
        self.grad_mean = grad_mean
        self.grad_var = grad_var
        self.reset()
    
    def reset(self):
        self.theta = self.init_theta
        self.w = self.init_w
    
    @property
    def params(self):
        return np.array([self.theta, self.w])

    @params.setter
    def params(self, params):
        self.theta, self.w = params
        
    def draw_action(self, s):
        mu = self.mean(s, self.theta)
        std = math.sqrt(self.var(s, self.w))
        return np.random.normal(mu, std)

    def gradient_log(self, s, a):
        mu = self.mean(s, self.theta)
        s2 = self.var(s, self.w)
        grad_theta = (a - mu) / s2 * self.grad_mean(s, self.theta)
        grad_sigma = ((a - mu) ** 2 - s2) / (s2 ** 2) * self.grad_var(s, self.w)
        return np.array([grad_theta, grad_sigma])
    
```

```python
from abc import abstractmethod

class Stepper:
    
    @abstractmethod
    def reset(self):
        pass
    
    @abstractmethod
    def update(self, gt):
        pass
```

```python
class ConstantStepper(Stepper):
    
    def __init__(self, alpha):
        self.alpha = alpha
    
    def reset(self):
        pass
    
    def update(self, gt):
        return self.alpha * gt
```

```python
class DiscountedStepper(Stepper):
    
    def __init__(self, alpha, discount):
        self.initial_alpha = alpha
        self.discount = discount
        self.reset()
    
    def reset(self):
        self.alpha = self.initial_alpha
    
    def update(self, gt):
        #print(self.alpha)
        delta = self.alpha * gt
        #self.alpha *= self.discount
        return delta
    
```

```python
class AdamStepper(Stepper):
    
    def __init__(self, alpha = 0.1, b1 = 0.9, b2 = 0.999, e = 1e-8):
        self.alpha = alpha
        self.b1 = b1
        self.b2 = b2
        self.e = e
        self.reset()
        
    def reset(self):
        self.t = 1
        self.m = 0
        self.v = 0
        pass
    
    def update(self, gt):
        self.m = self.b1 * self.m + (1 - self.b1) * gt
        self.v = self.b2 * self.v + (1 - self.b2) * (gt * gt)
        self.m_corrected = self.m / (1 - self.b1 ** self.t)
        self.v_corrected = self.v / (1 - self.b2 ** self.t)
        self.t += 1
        return self.alpha / (np.sqrt(self.v_corrected) + self.e) * self.m_corrected
        
```

```python
class REINFORCE:
    
    def trajectory(self):
        
        s = self.env.reset()
        
        trajectory_grad = 0
        trajectory_reward = 0
        current_discount = 1
        
        for t in range(self.T):

            a = self.policy.draw_action(s)
            trajectory_grad += self.policy.gradient_log(s, a)
            s, r, done, _  = self.env.step(a)

            trajectory_reward += current_discount * r
            current_discount *= self.discount
            
            if done:
                break

        return trajectory_reward, trajectory_grad
            
    
    def step(self):
        
        average_reward = 0
        average_gradient = 0
        for n in range(self.N):
            
            trajectory_reward, trajectory_gradient = self.trajectory()
            
            sample_gradient = trajectory_reward * trajectory_gradient
            
            average_gradient += (sample_gradient - average_gradient) / (n + 1)
            average_reward += (trajectory_reward - average_reward) / (n + 1)
        
        policy.params = policy.params + self.stepper.update(average_gradient)
        
        return average_reward, policy.params
        
    def __init__(self, env, N, T, discount, policy, stepper, keep_hist = False):
        self.env = env
        self.N = N
        self.T = T
        self.discount = discount
        self.policy = policy
        self.stepper = stepper
        self.keep_hist = keep_hist
        self.reset()
        
    def reset(self):
        self.policy.reset()
        self.stepper.reset()
        self.params_hist = []
        self.reward_hist = []
        
    def run(self, n_itr):
        
        for i in tqdm(range(n_itr)):
            
            reward, params = self.step()
            
            if self.keep_hist:
                self.params_hist.append(params)
                self.reward_hist.append(reward)
```

```python
env = lqg1d.LQG1D(initial_state_type='random')

mean = lambda s, theta: np.asscalar(s) * theta
var = lambda s, w: 0.5 ** 2

grad_mean = lambda s, theta: np.asscalar(s)
grad_var = lambda s, w: 0

policy = GaussianPolicy(0, 0, mean, var, grad_mean, grad_var)

discount = 0.9
K = env.computeOptimalK(discount)
cov = 0.001
```

```python
def plot_rewards(reward_histories, labels, output=None, it_factor=1):
    plt.figure(figsize=(6, 4))
    for reward_history, label in zip(reward_histories, labels):
        iterations = np.arange(len(reward_history)) * it_factor
        plt.plot(iterations, reward_history, label=label)
    plt.xlabel("Number of iterations")
    plt.ylabel("Average return")
    plt.xlim(0, len(reward_histories[0]) * it_factor)
    plt.legend(loc='lower right')
    plt.tight_layout()
    if output is not None:
        plt.savefig("../imgs/" + output)
    plt.show()
    
def plot_param(params_histories, labels, output=None, it_factor=1):
    plt.figure(figsize=(6, 4))
    for params_history, label in zip(params_histories, labels):
        iterations = np.arange(len(params_history[:, 0])) * it_factor
        plt.plot(iterations, params_history[:, 0], label=label)
    plt.axhline(np.asscalar(K), color="goldenrod", dashes=[2, 2])
    plt.xlabel("Number of iterations")
    plt.ylabel("Average mean parameter (theta)")
    plt.xlim(0, len(params_histories[0]) * it_factor)
    plt.legend(loc='upper right')
    plt.tight_layout()
    if output is not None:
        plt.savefig("../imgs/" + output)
    plt.show()
```

# Variation of the learning rate (constant stepper)

```python
n_params = 2

def evaluate_steppers(steppers, n_avg = 5, n_itr = 100, N=100, T=100, discount=0.9):
    
    reward_histories = np.zeros((len(steppers), n_itr))
    params_histories = np.zeros((len(steppers), n_itr, n_params))

    for i, stepper in enumerate(steppers):

        reward_history = np.zeros(n_itr)
        params_history = np.zeros((n_itr, n_params))
        
        succ_avg = 0
        
        for j in range(n_avg):

            stepper.reset()
            policy.reset()

            reinforce = REINFORCE(env, N, T, discount, policy, stepper, keep_hist=True)
            reinforce.run(n_itr)

            # Threshold to only keep the value that have converged
            if np.min(reinforce.reward_hist) > -1000:
                succ_avg += 1
                reward_history += np.array(reinforce.reward_hist)
                params_history += np.array(reinforce.params_hist)

        reward_histories[i] = reward_history / succ_avg
        params_histories[i] = params_history / succ_avg

    return reward_histories, params_histories
```

```python
learning_rates = [3e-7, 1e-6, 3e-6, 1e-5, 3e-5]
steppers = [ConstantStepper(learning_rate) for learning_rate in learning_rates]
constant_reward_histories, constant_params_histories = evaluate_steppers(steppers, n_itr=100)
```

```python
labels = [("learning_rate=%e" % learning_rate) for learning_rate in learning_rates ]
plot_rewards(constant_reward_histories, labels, output="constant_return.svg")
plot_param(constant_params_histories, labels, output="constant_mean.svg")
```

# Impact of N

```python
def evaluate_N(stepper, N, total_itr=10000, n_avg = 5, T=100, discount=0.9):
    
    N = np.array(N)
    
    n_itr = (total_itr / N).astype(int)
    samp = int(total_itr / np.max(N))
    inner = n_itr / samp
    
    reward_histories = np.zeros((len(N), samp))
    params_histories = np.zeros((len(N), samp, n_params))

    for i, (Ni, n_itr_i) in enumerate(zip(N, n_itr)):

        reward_history = np.zeros(n_itr_i)
        params_history = np.zeros((n_itr_i, n_params))
        
        success = 0

        for j in range(n_avg):

            stepper.reset()
            policy.reset()

            reinforce = REINFORCE(env, Ni, T, discount, policy, stepper, keep_hist=True)
            reinforce.run(n_itr_i)
            
            # Threshold to only keep the value that have converged
            if np.min(reinforce.reward_hist) > -1000:
                success += 1
                reward_history += np.array(reinforce.reward_hist)
                params_history += np.array(reinforce.params_hist)
                
        ids = np.linspace(0, n_itr_i - 1, samp).astype(int)
        
        reward_histories[i] = reward_history[ids] / (success + 1e-8)
        params_histories[i] = params_history[ids] / (success + 1e-8)
        

    return reward_histories, params_histories
```

```python
stepper = AdamStepper(1e-2)
N_array = [1, 5, 10, 50, 100]
N_reward_histories, N_params_histories = evaluate_N(stepper, total_itr=10000, N=N_array)
```

```python
labels = [("N=%d" % N) for N in N_array]
plot_rewards(N_reward_histories, labels, output="N_return.svg", it_factor=100)
plot_param(N_params_histories, labels, output="N_mean.svg", it_factor=100)
```

# 2. Off-Policy Reinforcement Learning with Value Function Approximation

```python
class UniformPolicy:
    
    def __init__(self, lower_bound, upper_bound):
        self.lb = lower_bound
        self.ub = upper_bound
        self.reset()
    
    def reset(self):
        pass
        
    def draw_action(self, s):
        return np.random.uniform(self.lb, self.ub)
```

```python
class QModel:
    
    def __init__(self):
        pass

    @abstractmethod
    def reset(self):
        pass
    
    @abstractmethod
    def max_over_action(self, state):
        pass
    
    @abstractmethod
    def regression_udpate(states, actions, expected):
        pass
    
```

```python
class QPolicy:
    
    def __init__(self, Q):
        self.Q = Q
        
    def draw_action(self, state):
        return self.Q.best_action(state)
```

```python
class LinearSecondOrderQModel(QModel):
    
    def __init__(self, theta, regul, actions):
        self.init_theta = theta
        self.regul = regul
        self.all_actions = actions
        self.reset()

    def reset(self):
        self.theta = self.init_theta
    
    def matrix(self, states, actions):
        A, S = np.meshgrid(actions, states)
        Z = np.dstack((
            A,
            A * S,
            S ** 2 + A ** 2
        ))
        return np.sum(Z * self.theta[None, None, :], axis=-1)
    
    def action_values(self, state):
        state = float(state)
        Z = np.vstack((
            self.all_actions,
            self.all_actions * state,
            self.all_actions ** 2 + state ** 2
        ))
        return np.dot(self.theta, Z)
    
    def max_over_action(self, state):
        Qstate = self.action_values(state)
        return np.max(Qstate)
    
    def best_action(self, state):
        Qstate = self.action_values(state)
        return self.all_actions[np.argmax(Qstate)]
    
    def regression_udpate(self, states, actions, expected):

        states = np.squeeze(states)
        Z = np.vstack((
            actions,
            states * actions,
            states ** 2 + actions ** 2
        ))
        
        I = np.dot(Z, Z.T) + self.regul * np.identity(3)
        new_theta = np.dot(np.linalg.inv(I), np.dot(Z, expected))
        self.theta = new_theta
    
```

```python
class FQI:
    
    def __init__(self, Q, discount = 0.95):
        self.Q = Q
        self.discount = discount
        self.reset()
    
    def reset(self):
        Q.reset()
    
    def step(self, dataset):
        
        rewards  = dataset["rewards"]
        states   = dataset["states"]
        actions  = dataset["actions"]
        n_states = dataset["next_states"]
        
        n = min(len(rewards), len(states),
                len(actions), len(n_states))
        
        # Compute Y
        Y = []
        for reward, n_state in zip(rewards, n_states):
            Y.append(reward + self.discount * Q.max_over_action(n_state))
        Y = np.array(Y)
       
        self.Q.regression_udpate(states[:n], actions[:n], Y[:n])

    def flatten_dataset(self, dataset):
        rewards  = []
        states   = []
        actions  = []
        n_states = []
        
        for trajectory in dataset:
            rewards.extend(trajectory['rewards'])
            states.extend(trajectory['states'])
            actions.extend(trajectory['actions'])
            n_states.extend(trajectory['next_states'])

        return {
            'rewards': np.array(rewards),
            'states': np.array(states),
            'actions': np.array(actions),
            'next_states': np.array(n_states)
        }
        
    def run(self, dataset, n_itr=1):
        
        dataset = self.flatten_dataset(dataset)
        
        for _ in tqdm(range(n_itr)):
            self.step(dataset)
            
    
    
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from utils import collect_episodes, estimate_performance

env = lqg1d.LQG1D(initial_state_type='random')

discount = 0.9
horizon = 50

actions = np.linspace(-8, 8, 20)
states = np.linspace(-10, 10, 20)

S, A = np.meshgrid(states, actions)

Q_fun = np.vectorize(lambda s, a: env.computeQFunction(s, a, K, cov, discount, 1))
Q_opt = Q_fun(S, A)

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(S.flatten(), A.flatten(), Q_opt.flatten())
ax.view_init(45, 70)
plt.tight_layout()
plt.savefig("../imgs/Q_optimal.svg")
plt.show()

#################################################################
# Collect the samples using the behavioural policy
#################################################################
# You should use discrete actions
beh_policy = UniformPolicy(-10.0, 10.0)

dataset = collect_episodes(env, n_episodes=100,
                                            policy=beh_policy, horizon=horizon)


Q = LinearSecondOrderQModel(0, 1e-4, actions)
fqi = FQI(Q, discount)
fqi.run(dataset, 100)

policy = QPolicy(fqi.Q)
Q_found = fqi.Q.matrix(states, actions)

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(S.flatten(), A.flatten(), Q_found.flatten())
ax.view_init(45, 70)
plt.tight_layout()
plt.savefig("../imgs/Q_fqi.svg")
plt.show()

J = estimate_performance(env, policy=policy, horizon=100, n_episodes=500, gamma=discount)
print('Policy performance: {}'.format(J))
plt.show()
```
